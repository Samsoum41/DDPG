> Structure de gestion des statistiques d'apprentissage :

```python
class MDP_statistics():
    """
        Statistics of Markov Decision Process over epochs on learning
    """

    def __init__(self, name : str = '', lengths : np.array = [], rewards : np.array = [], q_evols : np.array = []):
        self.name = str(name)
        self.lengths = np.asarray(lengths)
        self.rewards = np.asarray(rewards)
        self.q_evols = np.asarray(q_evols)

    def get_data_names(self) -> np.array(str):
        names = []
        attributes = vars(self) # Return all attributes names and values of self in a dict structure
        for name in list(attributes.keys())[1:]:
            if len(attributes[name]) != 0: # not empty
                names.append(name)
        return np.asarray(names)

    def __moving_average(self, values, width, mode) -> np.array:
        return np.convolve(values, np.ones(width), mode) / width

    def get_moving_averages(self, width : int, mode = 'valid') -> MDP_statistics:
        """
            :params     width           - bandwidth of the moving average
                        mode            - mode of the convolution used for moving average calculation
            :returns    MDP_statistics  - Moving average MDP_statistics 
        """
        return MDP_statistics(self.name,
                              self.__moving_average(self.lengths, width, mode),
                              self.__moving_average(self.rewards, width, mode),
                              self.__moving_average(self.q_evols, width, mode))
    
```

> Structure mère de Processus de décision Markovien pour l'apprentissage :

```python
class MDP:
    """
        Markov Decision Process
    """
    def __init__(self, learning_rate : float, discount_factor : float, seed : int = None):
        self.lr = learning_rate
        self.df = discount_factor
        # self.env.action_space.n returns the length of the action space
        self.q = defaultdict(lambda: np.zeros(self.env.action_space.n))
        self.policy = defaultdict(lambda: 0)
        self.env : gym.Env
        self.random_state = np.random.RandomState(seed)

    def _process_policy(self) -> None:
        """
            Process the policy by seeking the maximum action reward of the action-value function for each state
        """
        for state in self.q:
            self.policy[state] = int(np.argmax(self.q[state]))

    def reset_learning(self) -> None:
        self.q = defaultdict(lambda: np.zeros(self.env.action_space.n))
        self.policy = defaultdict(lambda: 0)

    def change_seed(self, seed : int) -> None:
        self.random_state = np.random.RandomState(seed)

    def _get_action(self, state, eps = 0.0, explore_flag = False) -> int:
        """
            :params     state           - current state of the MDP
                        eps             - rate of exploration action
                        explore_flag    - flag for forcing exploration
            :returns    action          - action number following the current version of self.q
        """
        if self.random_state.uniform() < eps or explore_flag:
            action = self.env.action_space.sample()
        else :
            action = int(np.argmax(self.q[state]))
        return action
    
    def _RL_algorithm_core(self, algo_name : str, updating_rule : 'function', epochs : int,
            eps_start = 0.0, eps_end = 0.0, eps_decay = 0.0, 
            explore_steps = 0) -> MDP_statistics:
        """
            This function is the core algorithm of any ReinforcementLearning_algorithm which we may call with any update rule.

            :params     updating_rule   - updating function of an RL algorithm (q-learning or sarsa)
                        epochs          - number of epochs
                        eps_start       - epsilon starting value for exploration
                        eps_end         - epsilon ending value for exploration
                        eps_decay       - epsilon decay over epochs
                        explore_steps   - Number of full exploration starting steps (burnin period)
            :returns    history         - history over learning (MDP_statistics object)
                            history.lengths     - length of game over epochs
                            history.rewards     - rewards of game over epochs
                            history.q_evols     - q evolutions over steps
        """
                
        # Statistics 
        lengths = []
        rewards = []
        q_evolutions = []

        eps = eps_start
        step = 0
        for _ in range(int(epochs)):
            ep_length = 0
            ep_reward = 0
            ep_q_evols = []
            state, _ = self.env.reset()
            terminated = False
            truncated = False
            while not (terminated or truncated):
                explore_flag = step < explore_steps
                step += 1
                action = self._get_action(state, eps, explore_flag)
                new_state, reward, terminated, truncated, _ = self.env.step(action)

                self.q[state][action], q_diff = updating_rule(state=state,
                                                                new_state=new_state,
                                                                action=action,
                                                                reward=reward,
                                                                terminated=terminated,
                                                                eps=eps,
                                                                explore_flag=explore_flag
                                                                )          

                state = new_state
                ep_reward += reward
                ep_length += 1
                ep_q_evols.append(q_diff)
            q_evolutions.append(np.mean(ep_q_evols))
            lengths.append(ep_length)
            rewards.append(ep_reward)
            if not explore_flag : # avoid lowering eps before the end of explore_steps
                eps = np.max([eps_end, eps - eps_decay])

        # Update of the state_value function with state_action_value function        
        self._process_policy()

        history = MDP_statistics(algo_name, lengths, rewards, q_evolutions)
        return history
    

    def __updating_rule_q_learning(self, state, new_state, action, reward, terminated, eps, explore_flag):
        max_next_q_value = (not terminated) * np.max(self.q[new_state])
        q_value = self.q[state][action]
        q_diff = reward + self.df * max_next_q_value - q_value
        q_value += self.lr*q_diff
        return q_value, q_diff

    def q_learning_greedy_algorithm(
            self, epochs : int,
            eps_start = 0.0, eps_end = 0.0, eps_decay = 0.0, 
            explore_steps = 0) -> MDP_statistics:
        """
            For q_learning algorithm without epsilon-greedy, let default values of eps related parameters
        """

        return self._RL_algorithm_core('q_learning_greedy, eps -> ' + str(eps_end), self.__updating_rule_q_learning, epochs,
            eps_start, eps_end, eps_decay, explore_steps)
    
    

    def __updating_rule_sarsa(self, state, new_state, action, reward, terminated, eps, explore_flag):
        new_action = self._get_action(new_state, eps, explore_flag)
        next_q_value = (not terminated) * self.q[new_state][new_action]
        q_value = self.q[state][action]
        q_diff = reward + self.df * next_q_value - q_value
        q_value += self.lr*q_diff
        return q_value, q_diff   
    
    def sarsa_algorithm(
            self, epochs : int,
            eps_start = 0.0, eps_end = 0.0, eps_decay = 0.0, 
            explore_steps = 0) -> MDP_statistics:
        """
            For sarsa algorithm without epsilon-greedy, let default values of eps related parameters
        """
        return self._RL_algorithm_core('sarsa, eps -> ' + str(eps_end), self.__updating_rule_sarsa, epochs,
            eps_start, eps_end, eps_decay, explore_steps)

    

    def evaluate_performance(self, nb_tests) -> MDP_statistics:
        """
        :params     nb_tests    - number of game for testing performances
        :returns    history     - statistics over testing
                            history.lengths        - length of game over tests
                            history.rewards        - rewards over tests
        """
        # Statistics 
        lengths = []
        rewards = []

        for _ in range(int(nb_tests)):
            test_length = 0
            test_reward = 0
            state, _ = self.env.reset()
            terminated = False
            truncated = False
            while not (terminated or truncated):
                action = self.policy[state]
                state, reward, terminated, truncated, _ = self.env.step(action)
                test_length += 1
                test_reward += reward
            lengths.append(test_length)
            rewards.append(test_reward)
        return MDP_statistics('Performance evaluation', lengths, rewards)
```

> Mise en place des outils pour l'analyse des résultats :

```python
class MDP_analyzer:
    """
        Toolbox class for human-analysis of a Markov Decision Process performances.
        This class manipulates MDP_statistics objects.
    """

    @classmethod
    def display_statistics(self, stats : MDP_statistics) -> None:
        data_names = stats.get_data_names()
        fig, ax = plt.subplots(ncols=len(data_names), figsize=(18, 5))
        for plot_number in range(len(data_names)):
            name : str = data_names[plot_number]
            data = getattr(stats, name) # Return a value of an attribute from a String
            ax[plot_number].set_title(name)
            ax[plot_number].plot(range(len(data)), data, label=stats.name)
            ax[plot_number].set_xlabel('epochs')
        # One global legend for all plots 
        handles, labels = plt.gca().get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        fig.legend(by_label.values(), by_label.keys())
    
    @classmethod
    def compare_statistics(self, *to_compare : MDP_statistics) -> None:
        data_names = to_compare[0].get_data_names()
        fig, ax = plt.subplots(ncols=len(data_names), figsize=(18, 5))
        for plot_number in range(len(data_names)):
            name : str = data_names[plot_number]
            ax[plot_number].set_title(name)
            for stats in to_compare:
                data = getattr(stats, name) # Return a value of an attribute from a String
                ax[plot_number].plot(range(len(data)), data, label=stats.name)
                ax[plot_number].set_xlabel('epochs')
        # One global legend for all plots 
        handles, labels = plt.gca().get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        fig.legend(by_label.values(), by_label.keys())

    @classmethod
    def test_algo_and_eps(self, mdp : MDP, epochs : int = 1e4, moving_average_scale : int = 500) -> tuple(np.array, np.array):
        hist = []
        hist.append(mdp.q_learning_greedy_algorithm(epochs, eps_start=1, eps_end=0.1, eps_decay=2/epochs, explore_steps=2000))
        mdp.reset_learning()
        hist.append(mdp.sarsa_algorithm(epochs, eps_start=1, eps_end=0.1, eps_decay=2/epochs, explore_steps=2000))
        mdp.reset_learning()
        hist.append(mdp.q_learning_greedy_algorithm(epochs, eps_start=1, eps_end=0.0, eps_decay=2/epochs, explore_steps=2000))
        mdp.reset_learning()
        hist.append(mdp.sarsa_algorithm(epochs, eps_start=1, eps_end=0.0, eps_decay=2/epochs, explore_steps=2000));
        hist_moving_average = []
        for element in hist:
            stat : MDP_statistics = element
            hist_moving_average.append(stat.get_moving_averages(width = moving_average_scale))
        return hist, hist_moving_average
    
    @classmethod
    def test_algo(self, mdp : MDP, epochs : int = 1e4,
                    eps_start : float = 1.0, eps_end = 0.0, 
                    eps_decay : float = -1, explore_steps : int = 2000, 
                    moving_average_scale : int = 500) -> tuple(np.array, np.array):
        if eps_decay == -1 :
            eps_decay = 2/epochs
        hist = []
        hist.append(mdp.q_learning_greedy_algorithm(epochs, eps_start, eps_end, eps_decay, explore_steps))
        mdp.reset_learning()
        hist.append(mdp.sarsa_algorithm(epochs, eps_start, eps_end, eps_decay, explore_steps))
        hist_moving_average = []
        for element in hist:
            stat : MDP_statistics = element
            hist_moving_average.append(stat.get_moving_averages(width = moving_average_scale))
        return hist, hist_moving_average
```

> Classe d'optimisation des hyperparamètres des méthodes :

```python
class MDP_optimizer():
    """
        Optimizer for Markov Decision Process Hyperparameter
    """

    @classmethod
    def best_hyperparameters(self, mdp : MDP, learning_rates : np.array, discount_factors : np.array, epochs : int) -> dict: 
        """
        :returns best_parameters    - Dictionnary with the best algorithm and the best hyperparameters 
        """

        maxi = -np.inf
        for df in discount_factors:
            for lr in learning_rates:
                mdp.lr = lr
                mdp.df = df
                mdp.reset_learning()

                rewards = mdp.q_learning_greedy_algorithm(epochs, eps_start=1, eps_end=0.1, eps_decay=2/epochs, explore_steps=2000).rewards
                last10p_mean = np.mean(rewards[int(epochs*9/10):])
                if last10p_mean > maxi:
                    maxi = last10p_mean
                    best_parameters = dict({'algorithm' : 'q_learning_greedy', 'learning_rate' : lr, 'discount_factor' : df})

                mdp.reset_learning()
                rewards = mdp.q_learning_greedy_algorithm(epochs, eps_start=1, eps_end=0.0, eps_decay=2/epochs, explore_steps=2000).rewards
                last10p_mean = np.mean(rewards[int(epochs*9/10):])
                if last10p_mean > maxi:
                    maxi = last10p_mean
                    best_parameters = dict({'algorithm' : 'q_learning_greedy eps->0', 'learning_rate' : lr, 'discount_factor' : df})

                mdp.reset_learning()
                rewards = mdp.sarsa_algorithm(epochs, eps_start=1, eps_end=0.1, eps_decay=2/epochs, explore_steps=2000).rewards
                last10p_mean = np.mean(rewards[int(epochs*9/10):])
                if last10p_mean > maxi:
                    maxi = last10p_mean
                    best_parameters = dict({'algorithm' : 'sarsa', 'learning_rate' : lr, 'discount_factor' : df})

                mdp.reset_learning()
                rewards = mdp.sarsa_algorithm(epochs, eps_start=1, eps_end=0.0, eps_decay=2/epochs, explore_steps=2000).rewards
                last10p_mean = np.mean(rewards[int(epochs*9/10):])
                if last10p_mean > maxi:
                    maxi = last10p_mean
                    best_parameters = dict({'algorithm' : 'sarsa eps->0', 'learning_rate' : lr, 'discount_factor' : df})
        return best_parameters
```

<u>Remarque</u> : Contrairement au blackjack, pour frozen lake, les algorithmes avec un epsilon qui tend vers 0 se démarquent avec des résultats bien meilleurs dès 50 000 epochs. Notons que dans ce jeu, la moyenne des rewards correspond à la fréquence des parties gagnées. Ici, comme on calcule une moyenne glissante sur 500 épisodes, il s'agit de la fréquence de victoire sans tomber dans un trou dans les 500 derniers épisodes. Contrairement au blackjack, le frozenlack classique avec une carte unique n'est pas sujet à l'aléatoire et converge vers une politique déterministe parfaite, peu importe l'algorithme.

**L'évolution de q_evols est particulière et se décompose en trois phases : (Elle indique l'évolution moyenne de la fonction state-action-value à chaque épisode)**

    - La fonction state_action_value q est initialisée à 0 pour tout état donc ses premières évolutions (temporal-difference) sont faibles. Elle croîent rapidement au départ du fait des mouvements aléatoires (Eps-greedy) qui font que de nombreuses valeurs de la fonction sont mises à jour.
    - Le epsilon tend vers 0 et les mouvements sont de moins en moins aléatoire, la politique commence à converger fortement vers un optimum et donc, comme le reward est nul pour une bonne action (hormis à la victoire où il vaut 1) et que les choix de la politique commencent à se répéter (bon choix), l'évolution de q diminue d'épisode en épisode.
    - Vers 50 000 épisodes, la politique a convergé et donc elle prend toujours les mêmes décisions, l'évolution de q devient quasiment nulle (elle serait nulle avec un discount_factor de 1)

**Quand epsilon ne tend pas vers 0, on voit que les algorithmes ont tendance à trouver un optimum local mais pas global, ci ce n'est moins bon que celui avec un epsilon qui tend vers 0. Ce minimum local correspond à un taux de réussite de 90%, soit exactement le taux de mouvements d'exploitation en fin d'apprentissage. (epsilon tend vers 0.1 donc le taux de mouvement d'exploration tend vers 0.1%)**

## Pendulum

Dans la suite de ce notebook, on s'intéressera à la résolution du problème du [Pendulum](https://www.gymlibrary.dev/environments/classic_control/pendulum/) disponible sur la plateforme gymnasium d'OpenAI :

L'objectif est d'apprendre au pendule comment atteindre  la position verticale et se stabiliser. 

Initialement, nous nous sommes intéressés à l'algorithme REINFORCE. Pour cela, nous avons utilisé un modèle paramétrique présenté dans le livre 'REINFORCEMENT LEARNING' de S. Sutton et G. Barton. Dans la section 13.7, le livre présente un modèle d'apprentissage de densités gaussiennes avec des modèles d'espérance et de variance pour la politique. Nous n'avons pas réussi à obtenir un algorithme d'apprentissage efficace (permettant de réussir un épisode).

Pour résoudre ce problème, nous avons choisi d'utiliser d'autres algorithmes de renforcement par montée de gradient sur une politique paramétrique, les modèles Acteur-Critique. En particulier, l'algorithme de Deep Deterministic Gradient Policy (DDPG). 

## Algorithme DDPG

L'algorithme DDPG utilise un modèle Acteur-Critique qui permet d'apprendre sur des espaces d'action et d'état continues. La politique paramétrique est un réseau de neurone, l'acteur. Pour entrainer cette politique, nous utilisons un modèle paramétrique de la Q-value fonction, le critique. Ainsi, à chaque proposition d'action de l'acteur, le critique évalue la Q-value fonction pour l'état et l'action proposée pour déterminer si c'est un bon choix ou non. Réciproquement, le critique exploite la politique pour s'améliorer avec l'équation de Bellman, où l'action est prévue par l'acteur.

Ainsi, l'acteur et le critique s'améliorent réciproquement jusqu'à atteindre la politique et la Q-value fonction optimales.

DDPG utilise aussi 2 techniques importantes :

- L'utilisation d'un buffer stockant les anciennes étapes : On garde en mémoire tous les tuples (state, action, reward, next_state). Ensuite, on réalisera les optimisations de chacun des réseaux de neurones sur un échantillon de taille **batch_size** constante tirés uniformément dans ce buffer. On notera que le buffer n'est pas vidé entre les épisodes.
- L'utilisation d'un doublon pour chaque réseau de neurones : Si on utilise le même réseau de neurone pour  qu'il se mette à jour lui-même, il devient instable. Pour éviter cela, on utilise une copie de chaque réseau (acteur et critique) nommé target pour mettre à jour les modèles. Ensuite, les paramètres des réseaux targets sont mis à jour à chaque étape de l'algorithme par la formule $$ \theta_{target} = \tau\theta_{target} + (1 - \tau)\theta_{model} $$ où $\tau$ est proche de 1.

L'algorithme est présenté précisément dans l'article "Continuous control with deep reinforcement learning" (https://arxiv.org/abs/1509.02971).

## Bruit

Pour piloter l'exploration de notre agent et aussi améliorer sa stabilité, nous utiliserons du bruit pour perturber l'action de la politique à chaque fois. On peut utiliser plusieurs bruits. Ici nous en présenterons 3.

- Le bruit d'Ornstein-Uhlenbeck, solution initiale proposée avec l'algorithme DDPG dans le ARTICLE (Continuous control with deep reinforcement learning, 2015, https://arxiv.org/abs/1509.02971). En effet, le processus d'Ornstein-Uhlenbeck a pour propriété d'être gaussien à chaque instant et à générer des variables aléatoires corrélés. C'est aussi un processus de Markov. Initialement, l'algorithme DDPG ajoute une réalisation de ce processus à chaque action que le réseau de neurone action réalise.

```python
class OUActionNoise:
    def __init__(self, mean, std_deviation, theta=0.15, dt=1e-2, x_initial=None):
        self.theta = theta
        self.mean = mean
        self.std_dev = std_deviation
        self.dt = dt
        self.x_initial = x_initial
        self.reset()

    def __call__(self):
        # Formula taken from https://www.wikipedia.org/wiki/Ornstein-Uhlenbeck_process.
        x = (
            self.x_prev
            + self.theta * (self.mean - self.x_prev) * self.dt
            + self.std_dev * np.sqrt(self.dt) * np.random.normal(size=self.mean.shape)
        )
        # Store x into x_prev
        # Makes next noise dependent on current one
        self.x_prev = x
        return x

    def reset(self):
        if self.x_initial is not None:
            self.x_prev = self.x_initial
        else:
            self.x_prev = np.zeros_like(self.mean)
```

- Le bruit gaussien, simplement généré par des variables aléatoires gaussiennes i.i.d, est une autre façon de réaliser du bruit. On ajoute une réalisation d'une gaussienne à l'action suggéré par la politique de l'acteur.

- Enfin, une dernière approche plus récente serait de perturber l'action non pas après qu'elle ait été calculé mais directement dans le réseau de neurones. En effet, on peut perturber légèrement les poids du réseau de neurone de l'acteur avant de lui soumettre l'état **state**. L'action résultante serait ainsi bruitée différemment.

```python
from enum import Enum

class Bruit(Enum):
    Gaussien = 1
    OU = 2
```

### Buffer

Pour implémenter l'algorithme DDPG, il faut utiliser un buffer. Le buffer est un élément important et utile dans l'algorithme DDPG car il permet d'éviter les problèmes de corrélation entre les transitions, améliore la stabilité de l'apprentissage et permet de minimiser les effets de l'instabilité des mises à jour.

```python
class ReplayBuffer:
    def __init__(self, buffer_capacity=100000, batch_size=64):
        # Number of "experiences" to store at max
        self.buffer_capacity = buffer_capacity
        self.batch_size = batch_size
        # Its tells us num of times record() was called.
        self.buffer_counter = 0

        # Instead of list of tuples as the exp.replay concept go
        # We use different np.arrays for each tuple element
        self.state_buffer = np.zeros((self.buffer_capacity, num_states))
        self.action_buffer = np.zeros((self.buffer_capacity, num_actions))
        self.reward_buffer = np.zeros((self.buffer_capacity, 1))
        self.next_state_buffer = np.zeros((self.buffer_capacity, num_states))

    # Takes (s,a,r,s') obervation tuple as input
    def record(self, obs_tuple):
        index = self.buffer_counter % self.buffer_capacity
        self.state_buffer[index] = obs_tuple[0]
        self.action_buffer[index] = obs_tuple[1]
        self.reward_buffer[index] = obs_tuple[2]
        self.next_state_buffer[index] = obs_tuple[3]
        self.buffer_counter += 1

    # Return a batch from the buffer
    def getBatch(self):
        record_range = min(self.buffer_counter, self.buffer_capacity)
        batch_indices = np.random.choice(record_range, self.batch_size)
        
        state_batch = tf.convert_to_tensor(self.state_buffer[batch_indices])
        action_batch = tf.convert_to_tensor(self.action_buffer[batch_indices])
        reward_batch = tf.convert_to_tensor(self.reward_buffer[batch_indices], dtype = tf.float32)
        next_state_batch = tf.convert_to_tensor(self.next_state_buffer[batch_indices])

        return state_batch, action_batch, reward_batch, next_state_batch
```

### Acteur et Critique

On définit une classe Actor pour représenter l'acteur : 

```python
class Actor : 
    def __init__(self, lr, tau = 0.005, bruit = Bruit.Gaussien, std_dev = 0.25):
        self.lr = lr
        self.tau = tau
        self.model = self.make_actor()
        self.target = self.make_actor()
        self.bruit = bruit
        self.std_dev = std_dev
        self.optimizer = tf.keras.optimizers.Adam(lr)
        # Making the weights equal initially
        self.target.set_weights(self.model.get_weights())
        
        if bruit == Bruit.OU:
            self.noise_generator = OUActionNoise(mean=np.zeros(1), std_deviation=float(std_dev) * np.ones(1))
        elif bruit == Bruit.Gaussien: 
            self.noise_generator = lambda : np.array([np.random.randn()*self.std_dev])
        else: 
            raise Exception("Bruit non attendu")

    def make_actor(self):
        # Initialize weights between -3e-3 and 3-e3
        last_init = tf.random_uniform_initializer(minval=-0.003, maxval=0.003)

        inputs = layers.Input(shape=(num_states,))
        out = layers.Dense(256, activation="relu")(inputs)
        out = layers.Dense(256, activation="relu")(out)
        outputs = layers.Dense(1, activation="tanh", kernel_initializer=last_init)(out)

        # Our upper bound is 2.0 for Pendulum.
        outputs = outputs * upper_bound
        model = tf.keras.Model(inputs, outputs)
        return model
    
    def policy_parameter_space(self, state):
        return True
    
    def policy(self, state):
        tf_state = tf.expand_dims(tf.convert_to_tensor(state), 0)
        sampled_actions = tf.squeeze(self.model(tf_state))
        sampled_actions = sampled_actions.numpy() + self.noise_generator()

        # We make sure action is within bounds
        legal_action = np.clip(sampled_actions, lower_bound, upper_bound)
        return [np.squeeze(legal_action)]
    
    def update(self, batch, critic):
        state_batch, action_batch, reward_batch, next_state_batch = batch        
        with tf.GradientTape() as tape:
            actions = self.model(state_batch, training=True)
            critic_value = critic.model([state_batch, actions], training=True)
            actor_loss = -tf.math.reduce_mean(critic_value)

        actor_grad = tape.gradient(actor_loss, self.model.trainable_variables)
        self.optimizer.apply_gradients(
            zip(actor_grad, self.model.trainable_variables)
        )
        
    # This update target parameters slowly
    # Based on rate `tau`, which is much less than one.
    @tf.function
    def update_target(self):
        for (a, b) in zip(self.target.variables, self.model.variables):
            a.assign(b * self.tau + a * (1 - self.tau))
```

On définit une classe Critic pour représenter l'acteur : 

```python
class Critic:
    def __init__(self, lr, tau = 0.005):
        self.lr = lr
        self.tau = tau
        self.model = self.make_critic()
        self.target = self.make_critic()
        self.optimizer = tf.keras.optimizers.Adam(lr)

        # Making the weights equal initially
        self.target.set_weights(self.model.get_weights())


    def make_critic(self):
        # State as input
        state_input = layers.Input(shape=(num_states))
        state_out = layers.Dense(16, activation="relu")(state_input)
        state_out = layers.Dense(32, activation="relu")(state_out)

        # Action as input
        action_input = layers.Input(shape=(num_actions))
        action_out = layers.Dense(32, activation="relu")(action_input)

        # Both are passed through seperate layer before concatenating
        concat = layers.Concatenate()([state_out, action_out])

        out = layers.Dense(256, activation="relu")(concat)
        out = layers.Dense(256, activation="relu")(out)
        outputs = layers.Dense(1)(out)

        # Outputs single value for give state-action
        model = tf.keras.Model([state_input, action_input], outputs)

        return model
    
    def update(self, batch, actor):
        state_batch, action_batch, reward_batch, next_state_batch = batch
        # Training and updating Actor & Critic networks.
        # See Pseudo Code.
        with tf.GradientTape() as tape:
            target_actions = actor.target(next_state_batch, training=True)
            y = reward_batch + gamma * self.target(
                [next_state_batch, target_actions], training=True
            )
            critic_value = self.model([state_batch, action_batch], training=True)
            critic_loss = tf.math.reduce_mean(tf.math.square(y - critic_value))

        critic_grad = tape.gradient(critic_loss, self.model.trainable_variables)
        self.optimizer.apply_gradients(
            zip(critic_grad, self.model.trainable_variables)
        )
    # This update target parameters slowly
    # Based on rate `tau`, which is much less than one.
    @tf.function
    def update_target(self):
        for (a, b) in zip(self.target.variables, self.model.variables):
            a.assign(b * self.tau + a * (1 - self.tau))
```
